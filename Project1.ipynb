```python
#Importing the pandas Library
import pandas as pd
```

```python
#Reading the csv File and storing it in the DataFrame
input_file="/content/Titanic-Dataset.csv"
titanic_df=pd.read_csv(input_file)
```

```python
#Printing the number of rows and columns of the Dataset
print("The No of Rows is",titanic_df.shape[0])
print("The No of Columns is",titanic_df.shape[1])
```

```python
#Printing the columns of the DataFrame
print(titanic_df.columns)
```

```python
#Printing the data types of the columns
print(titanic_df.dtypes)
```

```python
#Printing the first 5 rows of the DataFrame using head()
print(titanic_df.head())
```

```python
#Printing the last 5 rows of the DataFrame using tail()
print(titanic_df.tail(5))
```

```python
#Printing the unique values in each column
print(titanic_df.nunique())
```

```python
#Printing the number of null values in each column
print(titanic_df.isnull().sum())
```

```python
#Check for duplicate rows in the Titanic dataset
print(titanic_df.duplicated().sum())
```

```python
#Calculating the mean of the "Age" column
age_mean=titanic_df["Age"].mean()
#Filling the missing values in the "Age" column with mean
titanic_df["Age"]=titanic_df["Age"].fillna(age_mean)
```

```python
#Calculating the mode of the "Cabin" column
cabin_mode=titanic_df["Cabin"].mode()[0]
#Filling the missing values of the "Cabin" column with the mode
titanic_df["Cabin"]=titanic_df["Cabin"].fillna(cabin_mode)
```

```python
#Calculating the Mode of the Embarked column
embarked_mode=titanic_df["Embarked"].mode()[0]
#Filling the missing values in the "Embarked" column with its mode
titanic_df["Embarked"]=titanic_df["Embarked"].fillna(embarked_mode)
```

```python
print(titanic_df.isna().sum())
```

```python
#Saving The Cleaned dataset to a CSV File
titanic_df.to_csv("Cleaned_titanic_dataset.csv")
```

**Visualizations to gain insights from the dataset**

```python
#Importing the Seaborn library
import seaborn as sns
#Importing the matplotlib Library
import matplotlib.pyplot as plt
#CountPlot to show the number of people who survived or not
sns.countplot(data=titanic_df,x="Survived")
#Adding a title to  the plot
plt.title("The number of people who survived or not ")
#Displaying the plot
plt.show()
```

```python
#CountPlot to show the number of people who survived or not Gender Wise
sns.countplot(data=titanic_df, x='Survived', hue='Sex')
#Adding a title to  the plot
plt.title("The number of people who survived or not Gender Wise ")
#Displaying the plot
plt.show()
```

```python
#CountPlot to show the number of people who survived or not Class Wise
sns.countplot(data=titanic_df,x="Survived",hue="Pclass")
#Adding the title to the plot
plt.title("The number of people who survived Class Wise")
#Displaying the plot
plt.show()
```

```python

# Count the number of passengers in each class
class_counts = titanic_df['Pclass'].value_counts().sort_index()

# Define labels for classes
labels = ['1st Class', '2nd Class', '3rd Class']

# Create the pie chart

plt.pie(class_counts, labels=labels, autopct='%1.1f%%')

# Adding a title to the chart
plt.title('Class Distribution of Titanic Passengers')

# Display the chart
plt.show()
```

```python
#Plotting a Histogram to Show Age Distribution of the Passengers
plt.hist(titanic_df["Age"],bins=10)
#Adding Title to the plot
plt.title("Histogram displaying the Age distribution of Passengers")
#Adding the xlabel and ylabel
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
#Displaying the plot
plt.show()
```
